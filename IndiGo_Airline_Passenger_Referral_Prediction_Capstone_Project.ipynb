In today’s fast-moving and highly competitive airline industry, companies are constantly searching for ways to improve the passenger experience and build lasting relationships with their customers. In a sector where customer loyalty can be hard to achieve and even harder to maintain, focusing on the customer journey has become a top priority.

One of the key areas IndiGo has started to focus on is passenger referrals. In simple terms, this means whether or not a passenger is likely to talk positively about the airline to their friends, family, or social media followers. These personal recommendations are extremely powerful. In fact, a satisfied customer who shares their good experience can influence many other potential travelers. This directly affects the airline’s brand reputation and market presence. That is why being able to predict which customers will become promoters — and which won’t — can help IndiGo take action to improve the experience in smart and meaningful ways.

All of these opinions, whether positive or negative, gave IndiGo an opportunity to listen and learn.

The richness of this customer feedback cannot be overstated. Reviews offer real, unfiltered insights into how passengers feel about different aspects of their journey. For example, when many passengers mention cabin crew behavior in their reviews — either as a highlight or a problem — that becomes a clear signal of what matters most to them. This kind of feedback is far more detailed and helpful than general survey ratings because it includes personal stories and detailed opinions.

If IndiGo can successfully identify which types of customers are more likely to recommend the airline — based on their travel habits, review language, or satisfaction scores — then the company can take proactive steps to keep those passengers happy. More importantly, they can also focus on passengers who are not satisfied and try to fix their problems before they stop using the airline or spread negative word of mouth.

The ultimate goal of this project is to move from reactive service to predictive improvement. Instead of waiting for passengers to complain or leave bad reviews, IndiGo can start making data-driven decisions that improve the overall flying experience. This could mean offering targeted upgrades, training staff in customer service, optimizing booking platforms, or tailoring offers based on traveler type. As a result, the airline not only earns more positive reviews and referrals but also builds stronger trust with its passengers.

In summary, by carefully analyzing reviews from over a decade and applying modern prediction tools, IndiGo is taking a smart, future-ready approach to customer satisfaction. Passenger referrals are not just a marketing metric — they are a window into customer emotions and expectations. And for a company that wants to lead in both value and loyalty, understanding those referrals is a powerful step forward.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Copy the original dataset
data = df.copy()

# Drop rows with missing target (recommended)
data = data.dropna(subset=['recommended'])

# Fill or drop other missing values (can be improved later)
data = data.dropna()

# Convert target column to binary (yes → 1, no → 0)
data['recommended'] = data['recommended'].map({'yes': 1, 'no': 0})

# Drop columns not useful for prediction
data.drop(['author', 'review_date', 'customer_review', 'route', 'date_flown'], axis=1, inplace=True)

# Convert categorical columns to numeric using LabelEncoder
cat_cols = ['airline', 'aircraft', 'traveller_type', 'cabin']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cat_cols:
    data[col] = le.fit_transform(data[col])

# Display updated dataset info
data.info()
````

* Encoded categories like airline, aircraft, and cabin into numbers.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1: Service Quality and Recommendations

#### 1.  Hypothesis Definition

Null Hypothesis (H₀): Cabin service quality has no relationship with recommendation.

Alternate Hypothesis (H₁): Cabin service quality is significantly related to recommendation.


#### 2. Appropriate Statistical Test: Chi-Square Test of Independence

````python
# Perform Statistical Test to obtain P-Value

# Step 1: Create a contingency table
contingency_table = pd.crosstab(df['cabin_service'], df['recommended'])

# Step 2: Apply Chi-Square Test
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output Results Clearly
print("Chi-Square Value:", round(chi2, 2))
print("P-Value:", p)
print("Degrees of Freedom:", dof)

# Decision Rule (Significance Level = 0.05)
alpha = 0.05

if p < alpha:
    print("\nReject the Null Hypothesis")
    print(" There IS a significant relationship between service quality and recommendation.")
else:
    print("\n Do NOT reject the Null Hypothesis")
    print(" There is NO significant relationship.")
````

##### Which statistical test have you done to obtain P-Value?

I used the Chi-Square Test of Independence to find the P-Value.

##### Why did you choose the specific statistical test?

Because both cabin_service and recommended are categorical variables (like levels and yes/no).

The Chi-Square Test is designed to check whether two categorical variables are related or not.

It works by comparing the observed values (what we actually see) with the expected values (what we’d see if they were unrelated).

This helps us decide if recommendation decisions are influenced by cabin service quality.

### Hypothetical Statement - 2: Traveller Type vs Value for Money

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): Traveller type does NOT affect the value for money rating.

Alternate Hypothesis (H₁): Traveller type DOES affect the value for money rating.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Step 1: Create a contingency table
table = pd.crosstab(df['traveller_type'], df['value_for_money'])

# Step 2: Apply Chi-Square Test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table)

# Output Results
print("Chi-Square Value:", round(chi2, 2))
print("Degrees of Freedom:", dof)
print("P-Value:", p)

# Decision Rule (Significance Level = 0.05)
alpha = 0.05

if p < alpha:
    print("\nReject the Null Hypothesis")
    print(" Traveller type significantly affects value for money perception.")
else:
    print("\n Do NOT reject the Null Hypothesis")
    print(" No strong link found between traveller type and value perception.")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test of Independence

##### Why did you choose the specific statistical test?

Both traveller_type and value_for_money are categorical. This test checks whether there's a statistical link between the type of traveller and how they feel about pricing. It compares actual vs expected counts for each group.

### Hypothetical Statement - 3: Overall Rating vs Recommendation

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): People who recommend and don’t recommend give similar overall ratings.

Alternate Hypothesis (H₁): People who recommend give higher overall ratings.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value : Independent Samples T-Test

# Step 1: Split the two groups
yes_group = df[df['recommended'] == 'yes']['overall'].dropna()
no_group = df[df['recommended'] == 'no']['overall'].dropna()

# Step 2: Run Independent Samples T-Test
from scipy.stats import ttest_ind
t_stat, p_val = ttest_ind(yes_group, no_group, equal_var=False)

print("T-Statistic:", round(t_stat, 2))
print("P-Value:", p_val)

alpha = 0.05

if p_val < alpha:
    print("\n Reject the Null Hypothesis")
    print(" There is a significant difference in ratings between those who recommend and those who don’t.")
else:
    print("\n Do NOT reject the Null Hypothesis")
    print(" No significant difference found.")
````

##### Which statistical test have you done to obtain P-Value?

Independent Samples T-Test


##### Why did you choose the specific statistical test?

The column overall is numerical (rating 1–10), and recommended has two groups (yes or no).
This test is perfect to check if the average rating differs significantly between two independent groups.
It helps prove whether people who recommend IndiGo truly feel more satisfied overall.

````python
# Select your features wisely to avoid overfitting
# Get correlation matrix
corr = reviews_df.corr(numeric_only=True)

# Set threshold
cutoff = 0.85

# Store pairs
high_corr = set()

# Find high correlation pairs
for i in range(len(corr.columns)):

    for j in range(i):

        if abs(corr.iloc[i, j]) > cutoff:
            col_a = corr.columns[i]
            col_b = corr.columns[j]
            high_corr.add((col_a, col_b))

# Show high correlation pairs
print("High Correlation Pairs:")
for a, b in high_corr:
    print(a, "<-->", b)

# Drop second feature in each pair
drop_cols = [b for (a, b) in high_corr]
reviews_df.drop(drop_cols, axis=1, inplace=True)

print("Dropped columns:", drop_cols)
````
